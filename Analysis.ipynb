## Column Wrangling 
Create new columns/modify existing ones

### Add and calculate a new `Revenue_per_Age` column

```python
df['Revenue_per_Age'] = df['Revenue'] / df['Customer_Age']

df['Revenue_per_Age'].head()
```

### Add and calculate a new `Calculated_Cost` column

Use this formula

$$ Calculated\_Cost = Order\_Quantity * Unit\_Cost $$

```python
df['Calculated_Cost'] = df['Order_Quantity'] * df['Unit_Cost']

df['Calculated_Cost'].head()
```

### Modify all `Unit_Price` values adding 3% tax to them

```python
df['Unit_Price'] *= 1.03

df['Unit_Price'].head()
```

```python
df.loc[(df['Age_Group'] == 'Adults (35-64)') & (df['Country'] == 'United States'), 'Revenue'].mean()
```

### Add and calculate a new `Calculated_Date` column then parse your column into a datetime object

```python
df['Calculated_Date'] = df[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)

df['Calculated_Date'] = pd.to_datetime(df['Calculated_Date'])

df['Calculated_Date'].head()
```
